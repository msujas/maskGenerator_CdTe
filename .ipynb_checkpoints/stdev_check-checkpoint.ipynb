```python
#removing cosmics
%matplotlib inline
nstdevs = 5
allDataSets2 = {}
for s in subdirs:
    

    dataset2 = np.empty(shape = allDataSets[s].shape, dtype = np.float32)
    stdev = np.nanstd(allDataSets[s],axis = 2,dtype = np.float32)
    median = np.nanmedian(allDataSets[s],axis = 2)
    for c in range(len(allDataSets[s][0,0])):
        print(s,c)
        array = allDataSets[s][:,:,c]
        
        dataset2[:,:,c] = np.where(array > median+nstdevs*stdev,np.nan,array)
        
    allDataSets2[s] = dataset2
    del dataset2
    del stdev
    del median
```

```python
import pyFAI
def clearPyFAI_header(file):
    x,y,e = np.loadtxt(file,unpack = True, comments = '#')
    np.savetxt(file,np.array([x,y,e]).transpose(), '%.6f')
imageFile = r'C:\Users\kenneth1a\Documents\beamlineData\a311189_pdf\01_RbCuF3_Q0p5\pdf\xrd\average/average_gainCorrected.cbf'
poniname = r'C:\Users\kenneth1a\Documents\beamlineData\a311189_pdf/Si_0_15tilt.poni'
poni = pyFAI.load(poniname)
array = fabio.open(imageFile).data
for m in maskDct:
    mask = maskDct[m]
    maskname = m.replace('.edf','')
    outfile = f'maskData/RbCuF3_{maskname}.xye'
    outfile_2d = outfile.replace('.xye','.edf')
    x,y,e = poni.integrate1d(data = array, filename = outfile,mask = mask,polarization_factor = 0.99,unit = '2th_deg',
                correctSolidAngle = True, method = 'bbox',npt = 5000, error_model = 'poisson', safe = False)
    poni.integrate2d(data = array, filename = outfile_2d,mask = mask,polarization_factor = 0.99,unit = '2th_deg',
                correctSolidAngle = True, method = 'bbox',npt_rad = 5000,npt_azim = 360, error_model = 'poisson', safe = False)
    clearPyFAI_header(outfile)
    qy = y*4*np.pi*np.sin(x*np.pi/360)/0.25509
    outfileq = outfile.replace('.xye','_q.xye')
    np.savetxt(outfileq,np.array([x,qy,e]).transpose())
```
